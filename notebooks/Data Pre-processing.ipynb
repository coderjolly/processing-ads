```python
import os
import pandas as pd
import torchvision.transforms as T
from torchvision import datasets
import torch.utils.data as td
import sys
sys.path.append('../src/')
from multilabel.loader import MultiLabelDataset
```

```python
input_size = 128
norm_arr=([0.4955, 0.4955, 0.4955], [0.2892, 0.2892, 0.2892])
transform_dict = {"train": 
                        T.Compose([
                        #T.ToPILImage(),
                        T.Resize(size=(input_size, input_size)),
                        T.RandomHorizontalFlip(),
                        #T.ColorJitter(contrast=0.5),
                        T.RandomAdjustSharpness(2),
                        T.RandomAutocontrast(),

                        T.ToTensor(),
                        T.Normalize(*norm_arr)]),
                    "test_val": 
                        T.Compose([
                        #T.ToPILImage(),
                        T.Resize(size=input_size),
                        T.ToTensor(),
                        T.Normalize(*norm_arr)])
                    }
```

```python
train_dataset = MultiLabelDataset(root=os.path.join("../data/processed/", "images", "train"),
                                        dataframe=train,
                                        transform=transform_dict["train"])
```

```python
data_loader_train = td.DataLoader(train_dataset,
                                      batch_size=4,
                                      shuffle=True,
                                      drop_last=False,
                                      num_workers=0,
                                      pin_memory=True)
```
